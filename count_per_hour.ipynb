# 시간변수 추가

```python
import pandas as pd

# Function to extract hour
def extract_hour(hhmissss):
    # Ensure the value has 8 digits
    hhmissss_str = str(hhmissss).zfill(8)
    # Extract the hour part
    hour = int(hhmissss_str[:2])
    return hour

# Apply the function to create a new 'Hour' column
Time_DTG_01['Hour'] = Time_DTG_01['OPT_HHMISSSS'].apply(extract_hour)
# Time_DTG_01 = Time_DTG_01[Time_DTG_01['Hour']!=0]  # Displaying the first few rows for demonstration purposes

Time_DTG_02['Hour'] = Time_DTG_02['OPT_HHMISSSS'].apply(extract_hour)
# Time_DTG_02 = Time_DTG_02[Time_DTG_02['Hour']!=0] 

Time_DTG_03['Hour'] = Time_DTG_03['OPT_HHMISSSS'].apply(extract_hour)
# Time_DTG_03 = Time_DTG_03[Time_DTG_03['Hour']!=0] 

Time_DTG_04['Hour'] = Time_DTG_04['OPT_HHMISSSS'].apply(extract_hour)
# Time_DTG_04 = Time_DTG_04[Time_DTG_04['Hour']!=0] 

Time_DTG_05['Hour'] = Time_DTG_05['OPT_HHMISSSS'].apply(extract_hour)
# Time_DTG_05 = Time_DTG_05[Time_DTG_05['Hour']!=0] 

Time_DTG_06['Hour'] = Time_DTG_06['OPT_HHMISSSS'].apply(extract_hour)
# Time_DTG_06 = Time_DTG_06[Time_DTG_06['Hour']!=0] 

Time_DTG_07['Hour'] = Time_DTG_07['OPT_HHMISSSS'].apply(extract_hour)
# Time_DTG_07= Time_DTG_07[Time_DTG_07['Hour']!=0] 

Time_DTG_08['Hour'] = Time_DTG_08['OPT_HHMISSSS'].apply(extract_hour)
# Time_DTG_08 = Time_DTG_08[Time_DTG_08['Hour']!=0] 

Time_DTG_09['Hour'] = Time_DTG_09['OPT_HHMISSSS'].apply(extract_hour)
# Time_DTG_09 = Time_DTG_09[Time_DTG_09['Hour']!=0] 

Time_DTG_10['Hour'] = Time_DTG_10['OPT_HHMISSSS'].apply(extract_hour)
# Time_DTG_10 = Time_DTG_10[Time_DTG_10['Hour']!=0] 

Time_DTG_11['Hour'] = Time_DTG_11['OPT_HHMISSSS'].apply(extract_hour)
# Time_DTG_11 = Time_DTG_11[Time_DTG_11['Hour']!=0] 

Time_DTG_12['Hour'] = Time_DTG_12['OPT_HHMISSSS'].apply(extract_hour)
# Time_DTG_12 = Time_DTG_12[Time_DTG_12['Hour']!=0] 
```

# DTG_01

```python
import pandas as pd

# Assuming df is your DataFrame
# df = ... (your DataFrame loading or creation here)

# Columns of interest
columns_of_interest = [
    "Speeding", "Long-term speeding", "Sudden acceleration", "Sudden start", "Sudden deceleration", "Sudden stop",
    "Sudden left turn", "Sudden right turn", "Sudden U-turn", "Sudden overtaking", "Sudden lane change"
]

# Grouping by 'Hour' and summing the values
counts_per_hour = Time_DTG_01.groupby('Hour')[columns_of_interest].sum()


# Reordering columns to move 'Speeding' to the front
column_order = ['Speeding'] + [col for col in counts_per_hour.columns if col != 'Speeding']
counts_per_hour_01 = counts_per_hour.reindex(columns=column_order)

counts_per_hour_01['Total_Sum'] = counts_per_hour_01.sum(axis=1)

# Now 'Speeding' will be the first column in the DataFrame

counts_per_hour_01
```
